```python
proj_UTMK = pyproj.Proj(init='epsg:5174')
proj_WGS84 = pyproj.Proj(init='epsg:4326')
```

```python
def transfrom_utmk_to_w84(df):
    return pd.Series(pyproj.transform(proj_UTMK, proj_WGS84, df['좌표정보(X)'], df['좌표정보(Y)']), index=['좌표정보(X)', '좌표정보(Y)'])
```

```python
yongin_open_medicine_loc[['x_w84_v2', 'y_w84_v2']] = yongin_open_medicine_loc.apply(transfrom_utmk_to_w84, axis=1)
```

```python
def tm_to_wgs84(x, y, from_crs='epsg:2097', to_crs='epsg:4326'):
    transformer = pyproj.Transformer.from_crs(from_crs, to_crs, always_xy=True)
    lon, lat = transformer.transform(x, y)
    return lat, lon
```

```python
yongin_open_medicine_loc[['Latitude', 'Longitude']] = yongin_open_medicine_loc.apply(lambda row: tm_to_wgs84(row['좌표정보(X)'], row['좌표정보(Y)']), axis=1, result_type='expand')
```

```python
def calculate_offset(lat_actual, lon_actual, lat_df, lon_df):
    lat_offset = lat_actual - lat_df
    lon_offset = lon_actual - lon_df
    return lat_offset, lon_offset

def apply_offset(df, lat_col, lon_col, lat_offset, lon_offset):

    df[lat_col] = df[lat_col] + lat_offset
    df[lon_col] = df[lon_col] + lon_offset
    return df
```

```python
def correct_coordinates(df, actual_lat, actual_lon, lat_col, lon_col):

    df_lat = df[lat_col].iloc[0]
    df_lon = df[lon_col].iloc[0]
    
    lat_offset, lon_offset = calculate_offset(actual_lat, actual_lon, df_lat, df_lon)
    
    df_adjusted = apply_offset(df, lat_col, lon_col, lat_offset, lon_offset)
    
    return df_adjusted
```

```python

def UTMtoGEO(easting, northing):
    p1 = pyproj.Proj(init="epsg:5174")
    p2 = pyproj.Proj(init="epsg:4326")
    lon, lat = pyproj.transform(p1, p2, easting, northing)
    return round(lat, 6), round(lon, 6)
```
